#Introduction

**Modern Portfolio Theory**


Modern portfolio theory (MPT), or mean-variance analysis, is a mathematical framework for assembling a portfolio of assets such that the expected return is maximized for a given level of risk. 

It is a formalization and extension of diversification in investing, the idea that owning different kinds of financial assets is less risky than owning only one type. Its key insight is that an asset's risk and return should not be assessed by itself, but by how it contributes to a portfolio's overall risk and return. It uses the variance of asset prices as a proxy for risk.

 **Background**

Economist Harry Markowitz introduced MPT in a 1952 essay, for which he was later awarded a Nobel Memorial Prize in Economic Sciences; see Markowitz model.

#Optimization Model

Now we'll define the objectivefunction (maximizing the returns- make as much money as possible!). Indicate the fraction/proportion of each stock we're investing in, multiplied by its average return.



```python
from pyomo.environ import *
m = ConcreteModel() #Model

# defining variables
# each one is a stock
m.DVA = Var(within=NonNegativeReals, bounds=(0,1)) #Stock DVA
m.BHPLF = Var(within=NonNegativeReals, bounds=(0,1)) #Stock BHPLF
m.AMZN = Var(within=NonNegativeReals, bounds=(0,1)) #Stock AMZN
m.MSFT = Var(within=NonNegativeReals, bounds=(0,1)) #Stock MSFT
m.BTC_USD = Var(within=NonNegativeReals, bounds=(0,1)) #Crypto BTC
m.ETH_USD = Var(within=NonNegativeReals, bounds=(0,1)) #Crypto ETH
m.USDC_USD = Var(within=NonNegativeReals, bounds=(0,1)) #Crypto USDC
m.BNB_USD = Var(within=NonNegativeReals, bounds=(0,1)) #Crypto BNB
```

**Constraints** 

Regardless of how much money you start with, the amount you put in your portfolio must be equal to'1' or 100 percent... you must invest all of your money.

```python
# declare objective
m.objective = Objective(expr =               
                        m.DVA*df_return[0] + 
                        m.BHPLF*df_return[1] + 
                        m.AMZN*df_return[2] +
                        m.MSFT*df_return[3] +
                        m.BTC_USD*df_return[4] +
                        m.ETH_USD*df_return[5] +
                        m.USDC_USD*df_return[6] + 
                        m.BNB_USD*df_return[7],
                        sense=maximize) # we want to MAX our returns!

# declare constraints
# 1. Sum of all proportions = 1
m.sum_proportions = Constraint(expr = m.DVA + m.BHPLF + m.AMZN + m.MSFT + m.BTC_USD + m.ETH_USD + m.USDC_USD + m.BNB_USD == 1)

# 2. Set maximum risk: WE ARE INCLUDING THIS CONSTRAINT NOW FOR TECHNICAL REASONS (see below), YOU CAN (and should!) IGNORE IT FOR NOW!
m.total_risk = Constraint(expr = m.DVA + m.BHPLF + m.AMZN + m.MSFT + m.BTC_USD + m.ETH_USD + m.USDC_USD + m.BNB_USD >= 0.0)
```

```python
# creating calculations table for calculate the risk
# 3. Calculate risk
def calc_risk(m):
  variables = m.DVA, m.BHPLF, m.AMZN, m.MSFT, m.BTC_USD, m.ETH_USD , m.USDC_USD , m.BNB_USD #For Stocks and Crypto currencies
  risk_exp = 0
  for i in range(len(variables)):
    for j in range(len(variables)):
      risk_exp += variables[i]*df_cov.at[tickers_list[i],tickers_list[j]]*variables[j]
  return risk_exp #risk

# We are going to use this expression to compute the risk
expr_risk = calc_risk(m)

# 3. Max risk should be less than 0.01
max_risk = 0.006
# Sequence of risk levels
risk_limits = np.arange(0.0001, max_risk, 0.00005) # take tiny steps
risk_limits #Risk levels
```

Sensitivity Report



```python
from pyomo.opt import SolverStatus, TerminationCondition

# updating risk contraint for each limit and then solving the problem
param_analysis = {} # key=risk, value =stock allocations 
returns = {} # key=risk, value = return
for r in risk_limits:
  # Remove constraint to avoid error messages
  m.del_component(m.total_risk)
  m.total_risk = Constraint(expr = expr_risk <= r)
  # run solver
  result = SolverFactory('bonmin', executable='/content/bonmin').solve(m)

  # If solution is not feasible, ignore this run
  if result.solver.termination_condition == TerminationCondition.infeasible:
    continue

  result = result.write()

  # store our allocation proportions
  param_analysis[r] = [m.DVA(), m.BHPLF(), m.AMZN(), m.MSFT(), m.BTC_USD(), m.ETH_USD() , m.USDC_USD() , m.BNB_USD()]
  # store our returns
  returns[r] =   m.DVA()*df_return[0] + m.BHPLF()*df_return[1] + m.AMZN()*df_return[2] + m.MSFT()*df_return[3] + m.BTC_USD()*df_return[4] + m.ETH_USD()*df_return[5] + m.USDC_USD()*df_return[6] + m.BNB_USD()*df_return[7]
```

```python
np.random.seed(2022)
iterations = 1000 #Looping of n times
return_val = 0 #Final returns
arr=[]
loss = 0
for i in range(0, iterations): #For loop to iterate
  simulations = np.random.multivariate_normal(bus, df_cov_new) #simulations value
  return_val = simulations[0]*asset_1 + simulations[1]*asset_2 + simulations[2]*asset_3 + simulations[3]*asset_4 #calculating the value
  arr.append(return_val) #appending the return value for each iteration
  if return_val < 0: #Counting of loss
    loss += 1
print("Number of times loss has occured:" ,loss) #print no of loss
print(sum(arr))
print("Percentage of losses incurred:", loss/iterations*100,"%") #probability
```

###Optimization Model-2

```python
m3 = ConcreteModel()


# defining variables
m3.DVA = Var(within=NonNegativeReals, bounds=(0,1))
m3.BHPLF = Var(within=NonNegativeReals, bounds=(0,1))
m3.AMZN = Var(within=NonNegativeReals, bounds=(0,1))
m3.MSFT = Var(within=NonNegativeReals, bounds=(0,1))
m3.BTC_USD = Var(within=NonNegativeReals, bounds=(0,1))
m3.ETH_USD = Var(within=NonNegativeReals, bounds=(0,1))
m3.USDC_USD = Var(within=NonNegativeReals, bounds=(0,1))
m3.BNB_USD = Var(within=NonNegativeReals, bounds=(0,1))
```

```python
# declare objective
m3.objective = Objective(expr =  m3.DVA*df_return[0] + 
                        m3.BHPLF*df_return[1] + 
                        m3.AMZN*df_return[2] +
                        m3.MSFT*df_return[3] + 
                        m3.BTC_USD*df_return[4] +
                        m3.ETH_USD*df_return[5] +
                        m3.USDC_USD*df_return[6] +
                        m3.BNB_USD*df_return[7],
                        sense=maximize) # yes, we want to MAX our returns!

# declare constraints
# 1. Sum of all porportions = 1
m3.sum_proportions = Constraint(expr = m3.DVA + m3.BHPLF + m3.AMZN + m3.MSFT + m3.BTC_USD + m3.ETH_USD + m3.USDC_USD + m3.BNB_USD  == 1)

# 2. Total_risk >0 
m3.total_risk = Constraint(expr = m3.DVA + m3.BHPLF + m3.AMZN + m3.MSFT + m3.BTC_USD + m3.ETH_USD + m3.USDC_USD + m3.BNB_USD  >= 0.0)


#3. perform interger programming to allocate 10% of the budget to each stock
m3.DVA_new= Var(domain=Binary)
m3.BHPLF_new = Var(domain=Binary)
m3.AMZN_new = Var(domain=Binary)
m3.MSFT_new = Var(domain=Binary)

#Constraints to put a lower boound on the stocks
m3.constraints = ConstraintList()
m3.constraints.add(m3.DVA >= 0.1*m3.DVA_new) 
m3.constraints.add(m3.BHPLF >= 0.1*m3.BHPLF_new) 
m3.constraints.add(m3.AMZN >= 0.1*m3.AMZN_new) 
m3.constraints.add(m3.MSFT >= 0.1*m3.MSFT_new)
m3.constraints.add(m3.DVA_new + m3.BHPLF_new + m3.AMZN_new + m3.MSFT_new >=2)
```

```python
# 5. Set maximum risk
def calc_risk():
  # this is a list of all of your investment allocations (proportion invested in each stock)
  # this is UPDATED as the model is cranking - see how we are referncing the decision variables?
  
  proportions = [m3.DVA,m3.BHPLF,m3.AMZN,m3.MSFT,m3.BTC_USD,m3.ETH_USD,m3.USDC_USD,m3.BNB_USD]

  risk_matrix = [] # an empty list  
  for index, col in enumerate(df_cov.columns):
    i = 0
    risk = []
    for row in df_cov[col]: #iteration
      risk.append(proportions[index]*row*proportions[i])
      i = i + 1
    risk_matrix.append(risk)
  risk_matrix = pd.DataFrame(risk_matrix)
  return risk_matrix.sum().sum() #rsik avlue

# We are going to use this expression to compute the risk
expr_risk = calc_risk()
max_risk_1 = 0.002
risk_limits_1 = np.arange(0.0001, max_risk_1, 0.00005) # take tiny steps
risk_limits_1
```

```python
# updating risk contraint for each limit and then solving the problem
param_analysis_2 = {} # key=risk, value =stock allocations 
returns2 = {} # key=risk, value = return

for r in risk_limits:
  m3.del_component(m3.total_risk)
  m3.total_risk = Constraint(expr = expr_risk <= r)

  # run solver
  result = SolverFactory('bonmin', executable='/content/bonmin').solve(m3)

  # If solution is not feasible, ignore this run
  if result.solver.termination_condition == TerminationCondition.infeasible:
    continue

  result = result.write()

  # store our allocation proportions
  param_analysis_2[r] = [m3.DVA(), m3.BHPLF() ,m3.AMZN(), m3.MSFT(), m3.BTC_USD(), m3.ETH_USD(), m3.USDC_USD(), m3.BNB_USD()]
  # store our returns
  returns2[r] = m3.DVA()*df_return[0] + m3.BHPLF()*df_return[1] + m3.AMZN()*df_return[2] + m3.MSFT()*df_return[3] + m3.BTC_USD()*df_return[4] + m3.ETH_USD()*df_return[5] + m3.USDC_USD()*df_return[6] + m3.BNB_USD()*df_return[7]
```
